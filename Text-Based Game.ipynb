```python
def start_adventure():
    """
    Starts the adventure by introducing the player to the story.
    """
    print("Welcome to the Text-Based Adventure Game!")
    print("You find yourself standing at the edge of a dark forest. You have two choices:")
    print("1. Enter the forest.")
    print("2. Walk along the edge of the forest.")
    
    choice = input("What do you do? (Enter 1 or 2): ").strip()
    if choice == '1':
        enter_forest()
    elif choice == '2':
        walk_along_edge()
    else:
        print("Invalid choice. Please enter 1 or 2.")
        start_adventure()

def enter_forest():
    """
    Handles the scenario where the player enters the forest.
    """
    print("\nYou step into the forest. It's dark, and the sounds of wildlife echo around you.")
    print("You come across a fork in the path:")
    print("1. Take the left path.")
    print("2. Take the right path.")
    
    choice = input("Which path do you choose? (Enter 1 or 2): ").strip()
    if choice == '1':
        meet_wise_owl()
    elif choice == '2':
        encounter_bear()
    else:
        print("Invalid choice. Please enter 1 or 2.")
        enter_forest()

def walk_along_edge():
    """
    Handles the scenario where the player walks along the forest edge.
    """
    print("\nYou walk along the edge of the forest, enjoying the sunshine.")
    print("You come across a small hut. The door is slightly ajar.")
    print("1. Enter the hut.")
    print("2. Continue walking.")
    
    choice = input("What do you do? (Enter 1 or 2): ").strip()
    if choice == '1':
        find_treasure()
    elif choice == '2':
        fall_into_river()
    else:
        print("Invalid choice. Please enter 1 or 2.")
        walk_along_edge()

def meet_wise_owl():
    """
    Scenario where the player meets a wise owl.
    """
    print("\nYou take the left path and encounter a wise owl.")
    print("The owl offers to share its wisdom or give you a magic feather.")
    print("1. Ask for wisdom.")
    print("2. Take the magic feather.")
    
    choice = input("What do you do? (Enter 1 or 2): ").strip()
    if choice == '1':
        print("\nThe owl shares its wisdom: 'Always trust your instincts.' You feel enlightened and leave the forest safely.")
        end_game("success")
    elif choice == '2':
        print("\nThe magic feather grants you the ability to fly! You soar out of the forest and feel victorious.")
        end_game("success")
    else:
        print("Invalid choice. Please enter 1 or 2.")
        meet_wise_owl()

def encounter_bear():
    """
    Scenario where the player encounters a bear.
    """
    print("\nYou take the right path and come face-to-face with a bear!")
    print("1. Try to run away.")
    print("2. Offer the bear some food.")
    
    choice = input("What do you do? (Enter 1 or 2): ").strip()
    if choice == '1':
        print("\nYou try to run, but the bear catches up to you. Game over!")
        end_game("failure")
    elif choice == '2':
        print("\nThe bear happily eats the food and lets you pass. You exit the forest safely.")
        end_game("success")
    else:
        print("Invalid choice. Please enter 1 or 2.")
        encounter_bear()

def find_treasure():
    """
    Scenario where the player finds a treasure in the hut.
    """
    print("\nYou cautiously enter the hut and find a chest filled with gold!")
    print("You are now rich and leave the forest triumphant.")
    end_game("success")

def fall_into_river():
    """
    Scenario where the player falls into a river.
    """
    print("\nYou continue walking and accidentally fall into a river.")
    print("You manage to swim to safety but lose your belongings. Game over!")
    end_game("failure")

def end_game(outcome):
    """
    Ends the game with a success or failure message.
    """
    if outcome == "success":
        print("\nCongratulations! You completed the adventure successfully!")
    elif outcome == "failure":
        print("\nBetter luck next time! The adventure didn't go as planned.")
    print("Thanks for playing!")
    exit()

if __name__ == "__main__":
    start_adventure()
```
